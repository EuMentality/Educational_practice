# <center> Центральная предельная теорема своими руками

В этом задании предстоит проверить работу центральной предельной теоремы. 
    
* Задача — оценить распределение выборочного среднего случайной величины при разных объёмах выборок. Для этого при трёх и более значениях n (например, 5, 10, 50) следует сгенерировать 1000 выборок объёма n и построить гистограммы распределений их выборочных средних. 
* Используя информацию о среднем и дисперсии исходного распределения, посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Обратите внимание

## <center> В качестве "любимого" распределения было выбрано бета - распределение. 

```python
# Импортируем необходимые библиотеки
from scipy.stats import gamma
import scipy.stats as sts
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
# генерируем выборку объема = 10000 из бета распределения 'beta(a, b)' с параметрами: a = 2, b = 2   
a = 2
b = 2
beta_rv = sts.beta(a,b) 
sample = beta_rv.rvs(1000) 

# По сгенерированной выборке строим нормированную гистограмму распределения
plt.hist(sample,bins = 25,density = True, label = 'sample hist')

# Теоретическая плотность распределения
x = np.linspace(0,1,1000)
pdf = beta_rv.pdf(x)
plt.plot(x,pdf, label = 'theoretical pdf')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc='upper right')
```

## <center> Оценка распределения выборочного среднего СВ при объемах : n = [5, 10, 50]

### n = 5

```python
n = 5
a = 2
b = 2
EX = a/(a+b) # мат. ожидание СВ из бета-распределения с параметрами a,b
DX = a*b/(a+b)**2 / (a + b + 1) # дисперсия СВ  из бета распределения с параметрами a,b

# Выборочное среднее имеет нормальное распределение с параметрами ( EX_mean, DX_mean) по центральной предельной теореме
EX_mean = EX
DX_mean = DX/n 
Sigma_mean = (DX_mean)**0.5

# генерируем 1000 выборок выборочного среднего объема n = 5 из бета распределения
mean = []
for i in range(1000):
    sample = beta_rv.rvs(n)
    avg_sample = (sample.sum())/n
    mean.append(avg_sample)
    
# Построение гистограммы массива выборочных средних
plt.hist(mean,bins = 25,density = True, label = 'mean hist (n = 5)')

# Построение теоретической плотности распределения выборочного среднего. уже используются выводы из ЦПТ : 
# распределение выборочного среднего сводится к нормальному
x = np.linspace(0,1,1000)
norm_rv = sts.norm(EX_mean, Sigma_mean)
pdf = norm_rv.pdf(x)
plt.plot(x,pdf, label = 'theoretical pdf (n = 5)')
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.legend(loc = 'upper right')
```
